```python
import pandas as pd 
white_house = pd.read_csv("2015_Report_to_Congress_on_White_House_Staff.csv")
white_house
```

```python
names_length = []
for i in range(len(white_house)):
    name = len(white_house.get_value(i,"Position Title"))
    names_length.append(name)
print(names_length)
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

#Create a histogram for all salary in the White House
fig1 = plt.figure()
ax = fig1.add_subplot(1,1,1)
ax.hist(white_house["Salary"])
ax.set(xlabel='Salary', ylabel='Frequency', title='Histogram of all Salaries in the White House')
plt.show()

#Create scatter plot between name length and salary for all employees
fig2 = plt.figure()
ax1 = fig2.add_subplot(1,1,1)
ax1.set(xlim=(0,160), ylim=(30000,180000), xlabel = 'Name Length', ylabel ="Salary", title='Relationship between Salary and Name Length')
ax1.scatter(names_length, white_house["Salary"])
plt.show()

#Calculating the correlation between name length and salary for all employees
from scipy.stats.stats import pearsonr
r, p_value = pearsonr(names_length, white_house["Salary"])
print('\nThe correlation value between name length and salary for all employees is %d' %r 
      + '. This correlation might imply that there is no correalation between the two variables.')

#Total salary that the White House pays to its employees
total_salary=sum(white_house['Salary'])
print('\nThe total salary that the White House pays to its employees is %d' %total_salary + '.')

#Highest and lowest paid staffers
min_sal = min(white_house["Salary"])
max_sal = max(white_house["Salary"])
min_staffers = white_house["Name"][white_house["Salary"]==min_sal]
max_staffers = white_house["Name"][white_house["Salary"]==max_sal]
print('\nThe lowest paid staffers in the White House are:\n %s' %min_staffers)
print('\nThe highest paid staffers in the White House are:\n %s' %max_staffers)

#Word that is most commonly used in 'Position Title'

dict_word = {}
for title in white_house["Position Title"]:
    words = title.split()
    for word in words:
        if word not in dict_word and word not in ['THE','The','the']:
            dict_word[word] = 1
        if word in dict_word and word not in ['THE','The','the']:
            dict_word[word] +=1
max_key = 0
max_value = 0
for key in dict_word:
    if dict_word[key] > max_value:
        max_key = key
        max_value = dict_word[key]
print('\nThe most commonly used word in position title is %s' %max_key + ' with the frequency of %d' %max_value)
```
